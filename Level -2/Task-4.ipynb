```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

#Loading & Reading the DataSet
df=pd.read_csv("C:\\Users\\hp\\Downloads\\Dataset .csv")
```

### Task4: Restaurant chains

a.) Identify if there are any restaurant chains present in the dataset.

```python
restaurant_chain = df.groupby("Restaurant Name").size().reset_index(name="OutletCount")
new = restaurant_chain[restaurant_chain["OutletCount"] > 1]
restaurant_chain = new.sort_values(by="OutletCount", ascending=False)
```

```python
print(restaurant_chain.head(10))
```

```python
# Plotting the top 5 restaurant chains
plt.bar(restaurant_chain["Restaurant Name"][:5], restaurant_chain["OutletCount"][:5])
plt.xlabel("Restaurant Chain")
plt.ylabel("Number of Outlets")
plt.title("Top 5 Restaurant Chains by Number of Outlets")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

b.) Analyze the ratings and popularity of different restaurant chains.

```python
# Group by restaurant name and calculate average rating
restaurant_ratings = df.groupby("Restaurant Name")["Aggregate rating"].mean().reset_index()
restaurant_ratings.rename(columns={"Aggregate rating": "AverageRating"}, inplace=True)
restaurant_ratings = restaurant_ratings.sort_values(by="AverageRating", ascending=False)
print(restaurant_ratings.head(10))
```

```python
restaurant_votes = df.groupby("Restaurant Name")["Votes"].sum().reset_index()
restaurant_votes.rename(columns={"Votes": "TotalVotes"}, inplace=True)
restaurant_votes = restaurant_votes.sort_values(by="TotalVotes", ascending=False)
print(restaurant_votes.head(10))
```

```python


top_rated = restaurant_ratings.head(5)

plt.bar(top_rated["Restaurant Name"], top_rated["AverageRating"])
plt.title("Top 5 Restaurant Chains by AVG Ratings")
plt.xlabel("Restaurant Chain")
plt.show()
```

```python
top_votes = restaurant_votes.head(5)

plt.bar(top_votes["Restaurant Name"], top_votes["TotalVotes"])
plt.xticks(rotation=45)
plt.ylabel("Total Votes")
plt.title("Top 5 Restaurant Chains by Total Votes")
plt.tight_layout()
plt.show()
```
