### Task -3 : Price Range Distribution

a.) Create a histgram or bar chart to visualize the distribution of price
ranges among the restaurants.

```python
df.hist(column="Price range")
plt.title("Distribution of Price Ranges")
plt.xlabel("Price Range")
plt.xlabel("ounts")
plt.show()
```

b.) Calculate the percentage of restaurants each price range category.

```python
price_range_counts = df['Price range'].value_counts()

# Total number of restaurants
total_restaurants = len(df)

# Calculate percentage
price_range_percent = (price_range_counts / total_restaurants) * 100

# Round and display
price_range_percent = price_range_percent.round(2)
print("Percentage of restaurants in each price range:")
print(price_range_percent)
```

```python
# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    price_range_percent,
    labels=price_range_percent.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['green', 'orange', 'red', 'blue']
)

plt.title('Percentage of Restaurants by Price Range')
plt.axis('equal')  # Equal aspect ratio makes the pie chart circular
plt.show()
```
