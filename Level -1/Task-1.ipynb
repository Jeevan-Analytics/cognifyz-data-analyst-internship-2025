### Level 1, Task 1: Top Cusines

a.) Determine the top three most common cuisines in the dataset.

```python
# Get the top 3 most common cuisines
# Split the 'Cuisines' column by comma, explode into individual rows, then count
cuisines_count = df['Cuisines'].str.split(', ').explode().value_counts()

# Print all cuisine counts 
print(cuisines_count)

# Get the top 3 most common cuisines
top_cuisines = cuisines_count.head(3)
```

```python

# Define specific colors
colors = ['green', 'orange', 'red']

plt.figure(figsize=(8, 5))
top_cuisines.plot(kind='bar', color=colors, edgecolor='black')

plt.title('Top 3 Most Common Cuisines', fontsize=14)
plt.xlabel('Cuisine', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

b.) Calculate the percentage of restaurants that serve each of the top cuisines.

```python
# Total number of restaurants
total_restaurants = len(df)
print(total_restaurants)

# Get top cuisines
top_cuisines = cuisines_count.head()

# Calculate the number of unique restaurants serving each of the top cuisines
percentage_per_cuisine = (top_cuisines / total_restaurants) *100

# Display results
print("Percentage of restaurants serving for top_three cuisines:", percentage_per_cuisine )
```

```python
# Simple bar plot
percentage_per_cuisine.plot(kind='bar', color=['green', 'orange', 'red'])

plt.title('Top five Cuisines by Percentage')
plt.ylabel('Percentage (%)')
plt.xlabel('Cuisine')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
