```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import re

#Loading & Reading the DataSet
df=pd.read_csv("C:\\Users\\hp\\Downloads\\Dataset .csv")
```

### Task3 : Task: Price Range vs. Online Delivery andTable Booking

a.) Analyze if there is a relationship between theprice range and the availability of onlinedelivery and table booking.

To Analyze we need to "plot 2 charts to-gether"

```python
print(df.columns)

# Assuming correct column names:
x1 = df['Has Online delivery']
x2 = df['Has Table booking']
y = df['Average Cost for two']
```

Plot1 for Online Delivery with the Avg Cost                                          
Plot2 for Table Bookings with the Avg Cost

```python
plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
sns.boxenplot(x=x1, y=y)
plt.xlabel('Online Delivery')
plt.ylabel('Average Cost for two')
plt.title('Online delivery affected by the cost ')
plt.subplot(1,2,2)
sns.boxplot(x=x2,y=y)
plt.xlabel('Table booking')
plt.ylabel('Average cost for two')
plt.title('Table booking affected by the cost')
plt.tight_layout()
plt.show()
```

"Both Online delivery & Table bookings are negatively impacted by cost"

```python
print("High priced Restaurants which have Avg Cost more than 10,000 did'nt provise Online delivery as well as Table booking facilitied")
```

Let's Check the restaurants having Avg Cost Lessthan 10,000 

```python
x1 = 'Has Online delivery'
x2 = 'Has Table booking'

plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
sns.boxenplot(x=x1, y='Average Cost for two', data=df[df['Average Cost for two'] < 10000])
plt.title('Online delivery by cost')

plt.subplot(1,2,2)
sns.boxplot(x=x2, y='Average Cost for two', data=df[df['Average Cost for two'] < 10000])
plt.title('Table booking by cost')

plt.tight_layout()
plt.show()
```

#### Note :
The Online delivery facility is present only for those rerstaurants , where the Avg cost is less than 1000.              
                                                The Table booking facility is availible in restaurant with an Avg cost ranging from 1000 to 2000.
