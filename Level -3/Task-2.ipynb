```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import re

#Loading & Reading the DataSet
df=pd.read_csv("C:\\Users\\hp\\Downloads\\Dataset .csv")
```

### Task2 : Votes Analysis

a.) Identify the restaurants with the highest andlowest number of votes.

```python
highest_voted = df.groupby('Restaurant Name')['Votes'].sum()
highest_restaurant = highest_voted.idxmax()
highest_votes = highest_voted.max()

print(f"The Highest Voted Restaurant is: {highest_restaurant} with {highest_votes} votes")
```

```python
votes = df.groupby('Restaurant Name')['Votes'].sum()
votes_non_zero = votes[votes > 0]  # Exclude restaurants with 0 votes
lowest_voted_restaurant = votes_non_zero.idxmin()
lowest_votes = votes_non_zero.min()

print(f"The Lowest Voted Restaurant is: {lowest_voted_restaurant} with {lowest_votes} votes")
```

b.) Analyze if there is a correlation between thenumber of votes and the rating of arestaurant.

```python
# Calculate correlation
correlation =np.corrcoef(df['Votes'],df['Aggregate rating'])

Correlation=np.round(correlation,2)
print(correlation)


```

```python
plt.figure(figsize=(10,5), dpi=100)
sns.heatmap(correlation,annot=True,fmt=".2f",linewidths=.4)
plt.show()
```

```python
# Scatter plot with color based on rating
sns.scatterplot(data=df, x='Aggregate rating', y='Votes', hue='Aggregate rating', palette='viridis')

sns.regplot(data=df, x='Aggregate rating', y='Votes', scatter=False, color='red')

plt.title('Correlation between Aggregate Rating and Votes: 0.31')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Votes')
plt.grid(True)
plt.show()

print("As per the result, the correlation value = 0.31, which indicates a moderate positive correlation between these two variables.")
```
